```python
language_mapping = {
    'Malayali': 'South Indian',
    'Telugu': 'South Indian',
    'Tamil': 'South Indian',
    'Kannada': 'South Indian',
    'Korean': 'Foreign',
    'Spanish': 'Foreign',
    'English':'English',
    'Hindi':'Hindi',
    'Punjabi':'Punjabi'
}

data['Language_happy'] = data['Which language do you prefer to listen to when you are Happy ?'].map(language_mapping)
data['Language_sad'] = data['Which language do you prefer to listen to when you are Sad/Emotional ?'].map(language_mapping)
data['Language_angry'] = data['Which language do you prefer to listen to when you are Angry/Aggressive ?'].map(language_mapping)
data['Language_calm'] = data['Which language do you prefer to listen to when you are Calm/Relaxed ?'].map(language_mapping)
data['Language_anxious'] = data['Which language do you prefer to listen to when you are Anxious ?'].map(language_mapping)
data['Language_excited'] = data['Which language do you prefer to listen to when you are Excited/Energetic ?'].map(language_mapping)
data['Language_sentimental'] = data['Which language do you prefer to listen to when you are Sentimental/Nostalgic ?'].map(language_mapping)
data['Language_inspired'] = data['Which language do you prefer to listen to when you are feeling Inspired ?'].map(language_mapping)
data['Language_compassionate'] = data['Which language do you prefer to listen to when you are  Empathetic/Compassionate ?'].map(language_mapping)
data['Language_romantic'] = data['Which language do you prefer to listen to when you are feeling Romantic ?'].map(language_mapping)




```
